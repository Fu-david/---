```python
data = pd.read_excel('600085.xlsx')  # 替换为你的数据文件路径

obv = []
obv.append(0)  # 初始值为0

for i in range(1, len(data)):
    if data['Close'][i] > data['Close'][i-1]:
        obv.append(obv[i-1] + data['Volume'][i])
    elif data['Close'][i] < data['Close'][i-1]:
        obv.append(obv[i-1] - data['Volume'][i])
    else:
        obv.append(obv[i-1])

# 将OBV指标转换为Series对象
obv_series = pd.Series(obv, index=data.index)
obv_series
```

```python
data['OBV']=obv_series
data
```

```python
window = 20  # 移动窗口大小
std_dev = 2  # 标准差倍数

# 计算中轨线（BOLL）
data['BOLLMID'] = data['Close'].rolling(window).mean()

# 计算标准差
data['STD'] = data['Close'].rolling(window).std()

# 计算上轨线（UPPER）和下轨线（LOWER）
data['BOLLUPPER'] = data['BOLLMID'] + std_dev * data['STD']
data['BOLLLOWER'] = data['BOLLMID'] - std_dev * data['STD']

# 输出结果
data
```

```python
data.to_excel('600085.xlsx')
```
