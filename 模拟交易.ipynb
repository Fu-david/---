```python
money=0
numtick=10000 #买100手
price= df1['Close'].tolist()
yc=df1['yc'].tolist()
ycmoneytt=[316100]
improve=[0]
zsz=[316100]
for i in range(len(yc)-1):
    if yc[i] == 0:
        todayc=( price[i] - price[i+1] )* numtick
        money = money + todayc
    elif yc[i] == 1:
        todayc=( price[i+1] - price[i] )* numtick
        money = money + todayc
    cc=price[i+1] * numtick
    todaytt = cc + money
    #print(round(todayc))
    #print(round(money))
    zsz.append(round(cc))
    improve.append(round(money))
    ycmoneytt.append(round(todaytt))
```

```python
df1.insert(4, '策略增强收益',improve) 
df1.insert(5, '使用策略总市值',ycmoneytt) 
df1.insert(6, '仅持有股票总市值',zsz) 
df1
```
